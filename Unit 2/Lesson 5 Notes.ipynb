```python
my_list = iter(my_list)
type(my_list)
```

```python
next(my_list)
```

```python
for item in my_list:
    print(item)
```

```python
my_list = iter(my_list)
iterating = True

while iterating:
    try:
        item = next(my_list)
    except StopIteration:
        iterating = False
    else:
        print(item)
```
